```python
import matplotlib.pyplot as plt
from matplotlib.dates import DateFormatter
import numpy as np
import pandas as pd
import matplotlib.dates as mdates

Meteorological = pd.read_csv('Meteorological.csv')
Meteorological
```

```python
import matplotlib.pyplot as plt
from pandas.plotting import register_matplotlib_converters

register_matplotlib_converters()
%matplotlib inline

from matplotlib import rcParams
rcParams['figure.figsize'] = 60,20

x = Meteorological['Date']
y1 = Meteorological['Temp']
y2 = Meteorological['WS']

fig, ax1 = plt.subplots()

ax2 = ax1.twinx()

curve1 = ax1.plot(x, y1, label='Temp', color='g')
curve2 = ax2.plot(x, y2, label='WS', color='m')

ax1.set_xlabel("Date", fontsize=40)
ax1.set_ylabel("Temp °C", fontsize=40)

ax2.set_ylabel("WS m/s", fontsize=40)

ax1.tick_params(axis='x', labelsize=40)
ax1.tick_params(axis='y', labelsize=40)

ax2.tick_params(axis='y', labelsize=40)

formatter = mdates.DateFormatter("%Y-%m-%d")
ax1.xaxis.set_major_formatter(formatter)

locator = mdates.MonthLocator()
ax1.xaxis.set_major_locator(locator)

plt.grid(True)
plt.plot()
plt.show()
```
